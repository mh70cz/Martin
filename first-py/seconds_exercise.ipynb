```python
# Version A: Direct string concatenation
# We join all the day names together in one line using the + operator.
weekdays_a = "Monday " + "Tuesday " + "Wednesday " + "Thursday " + "Friday " + "Saturday " + "Sunday"
print(weekdays_a)  # This prints: Monday Tuesday Wednesday Thursday Friday Saturday Sunday


# Version B: Step-by-step concatenation with reassignment
# We start with the first day and then add each day one by-one.
weekdays_b = "Monday "                 # Start with Monday
weekdays_b = weekdays_b + "Tuesday "   # Add Tuesday to the string
weekdays_b = weekdays_b + "Wednesday " # Add Wednesday to the string
weekdays_b = weekdays_b + "Thursday "  # Add Thursday to the string
weekdays_b = weekdays_b + "Friday "    # Add Friday to the string
weekdays_b = weekdays_b + "Saturday "  # Add Saturday to the string
weekdays_b = weekdays_b + "Sunday"     # Finally, add Sunday (no extra space needed here)
print(weekdays_b)  # This prints the same complete week day string


# Version C: Using the shorthand += operator
# This is a quicker way to add (concatenate) day names to our string.
weekdays_c = "Monday "      # Start with Monday
weekdays_c += "Tuesday "    # Add Tuesday quickly
weekdays_c += "Wednesday "  # Add Wednesday quickly
weekdays_c += "Thursday "   # Add Thursday quickly
weekdays_c += "Friday "     # Add Friday quickly
weekdays_c += "Saturday "   # Add Saturday quickly
weekdays_c += "Sunday"      # Finally, add Sunday (no extra space needed here)
print(weekdays_c)  # This prints: Monday Tuesday Wednesday Thursday Friday Saturday Sunday
```
